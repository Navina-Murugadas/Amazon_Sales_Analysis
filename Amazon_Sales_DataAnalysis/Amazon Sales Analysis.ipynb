# Exploratory Data Analysis (EDA):

```python
df.columns
```

# Size:

```python
ax = sns.countplot(x='Size', data=df)
```

```python
ax = sns.countplot(x='Size',  data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

__From above graph, we can see that most of the people buy M-Size.__

# Groupby:

__The groupby() function in pandas is used to group data based on one or more columns in a DataFrame.__

```python
df.groupby(['Size'], as_index=False) ['Qty'].sum().sort_values(by='Qty', ascending=False)
```

```python
S_Qty = df.groupby(['Size'], as_index=False) ['Qty'].sum().sort_values(by='Qty', ascending=False)
sns.barplot(x='Size', y='Qty', data=S_Qty)
```

__From the above graph, we can see that most sold are of M-Size.__

# Courier Status:

```python
ax = sns.countplot(data=df, x='Courier Status', hue='Status')
plt.figure(figsize = (15,10))
plt.show()
```

__From the above graph, we can see that majority of the orders are shipped by means of "COURIER".__

```python
df['Size'].hist()
```

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize = (10,5))
plt.hist(column_data, bins=30, edgecolor='Black')
plt.xticks(rotation=90)
plt.show()
```

__From the above graph, we can see that most people buy "T-Shirts".__

```python
B2B_Check = df['B2B'].value_counts()
plt.pie(B2B_Check, labels=B2B_Check, autopct = '%1.1f%%')
plt.show()
```

```python
B2B_Check = df['B2B'].value_counts()
plt.pie(B2B_Check, labels=B2B_Check.index, autopct = '%1.1f%%')
plt.show()
```

__From the above pie chart, we can observe that the maximum buyers (99.2%) are "Retailers" & the rest (0.8%) are B2B buyers.__

```python
x_data = df['Category']
y_data = df['Size']

plt.scatter(x_data, y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

```python
plt.figure(figsize = (12,6))
sns.countplot(data=df, x='ship-state')
plt.xlabel('Ship-state')
plt.ylabel('Count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
top_10_states = df['ship-state'].value_counts().head(10)

plt.figure(figsize = (12,6))
sns.countplot(data=df[df['ship-state'].isin(top_10_states.index)], x='ship-state')
plt.xlabel('Ship-state')
plt.ylabel('Count')
plt.title('Distribution of State')
plt.xticks(rotation=45)
plt.show()
```

__From the above graph, we can observe that most of the buyers are from "Maharashtra".__

# Conclusion:
__This Data Analysis reveals that the Amazon has a significant customer base in "MAHARASHTRA".  Amazon mainly serves "RETAILERS", fulfils orders mostly through "COURIER" and experiences high demand for "M-Size" "T-Shirts".__
